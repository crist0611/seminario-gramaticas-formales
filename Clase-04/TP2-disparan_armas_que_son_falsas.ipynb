```python
import nltk
import re
import os, sys
import matplotlib
```

```python
def sr_parser(sentence, grammar):                      
    print(grammar)                                     
    sentence = sentence.lower()                        
    if sentence.endswith('.'):                         
        sent = re.sub('\.',' ',sentence)               
    else:                                              
        sent = sentence                                
    sent = sent.split()                                
    sr_parser = nltk.ShiftReduceParser(grammar)        
    for tree in sr_parser.parse(sent):                 
        print(tree)                                    
        return(tree)                                   
```

```python
oracion3 = 'disparan armas que son falsas'                                  
grammar = nltk.data.load('gramaticas/gramatica_oracion_3.cfg')      
sr_parser(oracion3, grammar)
```
